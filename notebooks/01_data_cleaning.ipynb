```python

# CELL 1: Import Libraries


import pandas as pd               # for loading and manipulating data
import numpy as np                # for numerical operations
import matplotlib.pyplot as plt   # for plotting
import os                         # for file path handling

# This makes sure our charts appear inside the notebook
%matplotlib inline

print(" libraries imported successfully!")
```

```python

# CELL 2: Load All Datasets




DATA_PATH = "../data/"

# Load all CSV files into pandas DataFrames
# A DataFrame is like an Excel spreadsheet inside Python

carbon_prices   = pd.read_csv(DATA_PATH + "carbon_prices_daily.csv")
climate_events  = pd.read_csv(DATA_PATH + "climate_events.csv")
co2_emissions   = pd.read_csv(DATA_PATH + "co2_emissions_yearly.csv")
energy_mix      = pd.read_csv(DATA_PATH + "energy_mix_yearly.csv")
temp_anomaly    = pd.read_csv(DATA_PATH + "temperature_anomaly_monthly.csv")

print(" Datasets loaded successfully!")
print(f"   carbon_prices  : {carbon_prices.shape[0]:,} rows, {carbon_prices.shape[1]} columns")
print(f"   climate_events : {climate_events.shape[0]:,} rows, {climate_events.shape[1]} columns")
print(f"   co2_emissions  : {co2_emissions.shape[0]:,} rows, {co2_emissions.shape[1]} columns")
print(f"   energy_mix     : {energy_mix.shape[0]:,} rows, {energy_mix.shape[1]} columns")
print(f"   temp_anomaly   : {temp_anomaly.shape[0]:,} rows, {temp_anomaly.shape[1]} columns")
```

```python

# CELL 3A: Explore Carbon Prices


print("=" * 50)
print("CARBON PRICES DATASET")
print("=" * 50)

# .head() shows the first 5 rows
print("\n--- First 5 rows ---")
print(carbon_prices.head())

# .info() shows column names, data types, and missing values
print("\n--- Column Info ---")
print(carbon_prices.info())

# .describe() shows statistics like min, max, average
print("\n--- Statistics ---")
print(carbon_prices.describe())
```

```python

# CELL 3B: Explore Climate Events


print("=" * 50)
print("CLIMATE EVENTS DATASET")
print("=" * 50)

print("\n--- First 5 rows ---")
print(climate_events.head())

print("\n--- Column Info ---")
print(climate_events.info())

print("\n--- All Event Types ---")
print(climate_events["event_type"].value_counts())
```

```python

# CELL 3C: Explore CO2 Emissions


print("=" * 50)
print("CO2 EMISSIONS DATASET")
print("=" * 50)

print("\n--- First 5 rows ---")
print(co2_emissions.head())

print("\n--- Column Info ---")
print(co2_emissions.info())

print("\n--- Countries in dataset ---")
print(co2_emissions["country"].unique())

print("\n--- Year range ---")
print(f"From {co2_emissions['year'].min()} to {co2_emissions['year'].max()}")
```

```python

# CELL 3D: Explore Energy Mix


print("=" * 50)
print("ENERGY MIX DATASET")
print("=" * 50)

print("\n--- First 5 rows ---")
print(energy_mix.head())

print("\n--- Column Info ---")
print(energy_mix.info())

print("\n--- Statistics ---")
print(energy_mix.describe())
```

```python

# CELL 3E: Explore Temperature Anomaly


print("=" * 50)
print("TEMPERATURE ANOMALY DATASET")
print("=" * 50)

print("\n--- First 5 rows ---")
print(temp_anomaly.head())

print("\n--- Column Info ---")
print(temp_anomaly.info())

print("\n--- Regions in dataset ---")
print(temp_anomaly["region"].unique())
```

```python

# CELL 4A: Clean Carbon Prices



# Convert the 'date' column from text to actual date format
# This lets us do time-based analysis later
carbon_prices["date"] = pd.to_datetime(carbon_prices["date"])

# --- Check for missing values ---
print("Missing values in carbon_prices:")
print(carbon_prices.isnull().sum())

# --- Remove duplicate rows if any ---
before = len(carbon_prices)
carbon_prices = carbon_prices.drop_duplicates()
after = len(carbon_prices)
print(f"\nRemoved {before - after} duplicate rows")

# --- Sort by date ---
carbon_prices = carbon_prices.sort_values("date").reset_index(drop=True)

print("\n Carbon prices cleaned!")
print(carbon_prices.head())
```

```python

# CELL 4B: Clean Climate Events


# --- Fix date column ---
climate_events["date"] = pd.to_datetime(climate_events["date"])

# --- Check for missing values ---
print("Missing values in climate_events:")
print(climate_events.isnull().sum())

# --- Remove duplicates ---
before = len(climate_events)
climate_events = climate_events.drop_duplicates()
after = len(climate_events)
print(f"\nRemoved {before - after} duplicate rows")

print("\n Climate events cleaned!")
print(climate_events.head())
```

```python

# CELL 4C: Clean CO2 Emissions


# --- Check for missing values ---
print("Missing values in co2_emissions:")
print(co2_emissions.isnull().sum())

# --- Remove duplicates ---
before = len(co2_emissions)
co2_emissions = co2_emissions.drop_duplicates()
after = len(co2_emissions)
print(f"\nRemoved {before - after} duplicate rows")

# --- Make sure year is integer ---
co2_emissions["year"] = co2_emissions["year"].astype(int)

# --- Sort by country and year ---
co2_emissions = co2_emissions.sort_values(["country", "year"]).reset_index(drop=True)

print("\n CO2 emissions cleaned!")
print(co2_emissions.head())
```

```python

# CELL 4D: Clean Energy Mix


# --- Check for missing values ---
print("Missing values in energy_mix:")
print(energy_mix.isnull().sum())

# --- Fill missing percentage values with 0
# (if a country has no solar, it means 0%, not unknown)
percentage_columns = [
    "coal_pct", "oil_pct", "gas_pct", "nuclear_pct",
    "hydro_pct", "solar_pct", "wind_pct",
    "other_renewables_pct", "renewables_total_pct", "fossil_total_pct"
]
energy_mix[percentage_columns] = energy_mix[percentage_columns].fillna(0)

# --- Remove duplicates ---
before = len(energy_mix)
energy_mix = energy_mix.drop_duplicates()
after = len(energy_mix)
print(f"Removed {before - after} duplicate rows")

print("\n✅ Energy mix cleaned!")
print(energy_mix.head())
```

```python

# CELL 4E: Clean Temperature Anomaly

# --- Check for missing values ---
print("Missing values in temp_anomaly:")
print(temp_anomaly.isnull().sum())

# --- Fill missing co2_ppm values using forward fill ---
temp_anomaly["co2_ppm"] = temp_anomaly["co2_ppm"].ffill()

# --- Make sure year and month are integers ---
temp_anomaly["year"]  = temp_anomaly["year"].astype(int)
temp_anomaly["month"] = temp_anomaly["month"].astype(int)

# --- Remove duplicates ---
before = len(temp_anomaly)
temp_anomaly = temp_anomaly.drop_duplicates()
after = len(temp_anomaly)
print(f"Removed {before - after} duplicate rows")
print("\n Temperature anomaly cleaned!")
print(temp_anomaly.head())
```

```python

# CELL 5: Save Cleaned Datasets


# Create a folder for cleaned data
CLEAN_PATH = "../data/cleaned/"
os.makedirs(CLEAN_PATH, exist_ok=True)

# Save all cleaned dataframes as new CSV files
carbon_prices.to_csv(CLEAN_PATH  + "carbon_prices_clean.csv",  index=False)
climate_events.to_csv(CLEAN_PATH + "climate_events_clean.csv", index=False)
co2_emissions.to_csv(CLEAN_PATH  + "co2_emissions_clean.csv",  index=False)
energy_mix.to_csv(CLEAN_PATH     + "energy_mix_clean.csv",     index=False)
temp_anomaly.to_csv(CLEAN_PATH   + "temp_anomaly_clean.csv",   index=False)

print(" All cleaned datasets saved to data/cleaned/")
```
